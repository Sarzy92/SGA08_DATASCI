```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
yvals = range(len(df.model))
ax.barh(yvals, df.disp, align='center', alpha=0.4)
plt.yticks(yvals,df.model)
plt.barh(df.model, df.disp)
plt.xlabel('disp')
plt.ylabel('model')
plt.title('Mtcars, Model and Disp')
plt.tight_layout()
plt.show
```
