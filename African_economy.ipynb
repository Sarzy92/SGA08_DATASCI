```python
sns.distplot(Df['year'].dropna(), bins=15, kde=False)
plt.ylabel('Count')
plt.title('African Ecomony Over the years')
```

```python
plt.hist(Df['year'][(Df['banking_crisis'] == 'crisis')].dropna(), bins=7, label='crisis', histtype='stepfilled')
plt.hist(Df['year'][(Df['banking_crisis'] == 'no_crisis')].dropna(), bins=7, label='no_crisis', alpha=.7, histtype='stepfilled')
plt.xlabel('year')
plt.ylabel('Count')
plt.title('Overview of banking crisis over the years')
plt.legend()
```

```python
grouped_by_year = Df.groupby(['year', 'systemic_crisis', 'banking_crisis'])
grouped_by_year.size()
```

```python
sns.catplot('year', col='systemic_crisis', data=Df, kind='count', height=7, aspect=.8)
plt.subplots_adjust(top=0.9)
plt.suptitle('African Economy', fontsize=16)
```
